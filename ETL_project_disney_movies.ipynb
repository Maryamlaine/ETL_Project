### Table 2: 
source: IMDB website https://www.imdb.com/list/ls068561553/

Extract from IMDB to get movie duration, IMDB_rating, Number_of_reviews,

```python
executable_path = {'executable_path': 'chromedriver.exe'}
browser = Browser("chrome", **executable_path, headless=False)
```

```python
rating_list = []
title_list = []
duration_list = []
number_of_reviews = []

url = "https://www.imdb.com/list/ls068561553/"
browser.visit(url)


for x in range(1, 7):
    html = browser.html
    soup = bs(html, 'html.parser')
    contents = soup.find_all("div", class_='lister-item-content')

    for content in contents:
        print("----------------------")
        
        #extract movie title and imdb-ratings
        try:
            title = content.find('a').text
            print(f"Movie title: {title}")
            title_list.append(title)

            rating = content.find('span', class_="ipl-rating-star__rating").text
            print(f"Rating: {rating}")
            rating_list.append(rating)            
            
          
        except:
            print(f"Oops, rating not found! ")
            rating_list.append("")
            
        # extract movie durations    
        try:
            duration = content.find('span', class_="runtime").text
            print(f"duration: {duration}")
            duration_list.append(duration)
            
        except:
            print("Oops, duration not found!")
            duration_list.append("")
        
        #extract number of reviews
        try:
            vote = content.find("span" , {"name" : "nv"}).text
            print(f"number_of_reviews: {vote}")
            number_of_reviews.append(vote)
        
        except:
            print(f"Oops, number of reviews not found!")
            number_of_reviews.append("")
            
            
            
    browser.click_link_by_partial_text('Next')
    time.sleep(2)
    
```

```python
browser.quit()
```

```python
print(len(rating_list), len(title_list), len(duration_list), len(number_of_reviews))
```

```python
# create a review dataframe
rating_df = pd.DataFrame({'Title': title_list,
                         'Duration': duration_list,
                         'IMDB_rating': rating_list,
                         'Number_of_Reviews': number_of_reviews})
rating_df
```

### Table 3 & Table 4: 
source: IMDB website https://www.imdb.com/list/ls068561553/


Extract directors and actors of each Disney movie

```python
executable_path = {'executable_path': 'chromedriver.exe'}
browser = Browser("chrome", **executable_path, headless=False)
```

```python
title_list = []
director_list = []
actor_list = []

url = "https://www.imdb.com/list/ls068561553/"
browser.visit(url)


for x in range(1, 8):
    html = browser.html
    soup = bs(html, 'html.parser')
    contents = soup.find_all("div", class_='lister-item-content')
        
        
    for content in contents:
        print("----------------------")
        
        title = content.find('a').text
        print(f"Movie title: {title}")
        title_list.append(title)
        
        # find the directors and actors
        tag = content.find_all("p", class_="text-muted text-small")
        #get all of directors and voice actors
        staff_member = tag[1].text
        # remove Punctuation from text
        tag_list = ''.join([text for text in staff_member if text not in (',', '|', ":")])
        # remove space and empty strings
        tag_list_1 = [i for i in tag_list.split("\n") if i]
        tag_list_2 = [i.rstrip().lstrip() for i in tag_list_1]
        new_tag_list = [i for i in tag_list_2 if i != ""]

        #extract director and actor names
        director = []
        actor = []
        for i in range(len(new_tag_list)):
            if new_tag_list[i] == "Stars" or new_tag_list[i] == "Star":
                star_index = i
        director = new_tag_list[1 : star_index]
        actor = new_tag_list[star_index + 1:len(new_tag_list)]
        
        director_list.append(director)
        actor_list.append(actor)

        string_director=", ".join(director)
        string_actor=", ".join(actor)
        print(f"directors: {string_director}")
        print(f"actors: {string_actor}")
          
            
    time.sleep(5)
    browser.click_link_by_partial_text('Next')

```

```python
browser.quit()
```

```python
print(len(title_list), len(director_list), len(actor_list))
```

```python
# make sure that directors belong to their corresponding movies
new_title_list=[]
new_director_list=[]

for i in range(len(director_list)):
    #count number of directors
    repet_times=len(director_list[i])
    
    for time in range(repet_times):
        new_title_list.append(title_list[i])
    
    for each in director_list[i]:
        new_director_list.append(each)
```

#### Director table

```python
# create a dataframe
director_df = pd.DataFrame({"title": new_title_list,
                           "director": new_director_list})
director_df
```

```python
#make sure actors belong to their corresponding movies
new_title_list_2=[]
new_actor_list=[]

for i in range(len(actor_list)):
    #count number of actors
    repet_times=len(actor_list[i])
    
    for time in range(repet_times):
        new_title_list_2.append(title_list[i])
    
    for each in actor_list[i]:
        new_actor_list.append(each)
```

```python
print(len(new_title_list_2), len(new_actor_list))
```

#### Actor table

```python
actor_df = pd.DataFrame({"title": new_title_list_2,
                        "cast": new_actor_list})
actor_df
```

# Step two: Transform

### Table 1: box office table

```python
disney_df['income_id'] = [i for i in range(1, len(disney_df) + 1)]
box_df = disney_df.copy()
box_df = box_df[['income_id', 'total_gross']]
box_df.columns = ['income_id', 'gross_income']
box_df.head()
```

### Table two: movie table

```python
# movie table
movie_df = pd.merge(disney_df, 
                   rating_df,
                   left_on = 'movie_title',
                   right_on= 'Title')

new_movie_df = movie_df.copy()

# fill na
new_movie_df['genre'].fillna("No Genre", inplace = True)
new_movie_df['Duration'].fillna("No Duration", inplace = True)

#drop duplicates
new_movie_df.drop_duplicates(["movie_title"], keep='last', inplace=True)

new_movie_df['movie_id'] = [i for i in range(1, len(new_movie_df) + 1 )]
new_movie_df = new_movie_df[['movie_id', "income_id", 'Title', 'Duration', 'genre', 'release_date']]

#change the column names
new_movie_df.columns = ['movie_id', "income_id", 'movie_title', 'duration', 'genre', 'release_date']
new_movie_df
```

```python
df_data = new_movie_df.merge(box_df, on='income_id', how='outer')
genre_sum = df_data.groupby(['genre']).sum()[['gross_income']].sort_values('gross_income', ascending=False)
genre_sum
```

### Table three: review table

```python
disney_review_table = pd.merge(new_movie_df, 
                               rating_df,
                               left_on = 'movie_title',
                               right_on= 'Title')
disney_review_table.head()
```

```python
# review table
new_rating_df = disney_review_table.copy()

#dropna
new_rating_df = new_rating_df[new_rating_df['IMDB_rating'] != ""]

#remove_duplicates
new_rating_df.drop_duplicates(["movie_title"], keep='last', inplace=True)


#set index
new_rating_df['rating_id'] = [i for i in range(1, len(new_rating_df) + 1)]
new_rating_df = new_rating_df[['rating_id','IMDB_rating', 'Number_of_Reviews', 'movie_id']]
new_rating_df.columns = ['review_id','imdb_rating', 'number_of_reviews', 'movie_id']


# change datatype for reviews column
new_rating_df['number_of_reviews'] = new_rating_df['number_of_reviews'].str.replace(',', '')
new_rating_df['number_of_reviews'] = new_rating_df['number_of_reviews'].astype(int)


new_rating_df
```

```python
# top 10 higher store movies
highest_score_movie = disney_review_table[['Title','IMDB_rating']].sort_values("IMDB_rating", ascending=False)
highest_score_movie.head(10)
```

```python
# top 10 higher review movies
highest_reviews_movie = disney_review_table[['Title', 'Number_of_Reviews']].sort_values("Number_of_Reviews", ascending=False)
highest_reviews_movie.head(10)
```

### Table four: Director table

```python
temp_director_df = director_df.copy()

#split name
temp_director_df.loc[temp_director_df['director'].str.split().str.len() == 2, 'last name'] = temp_director_df['director'].str.split().str[-1]

#dropna
temp_director_df = temp_director_df[temp_director_df['last name'].notna()]

#remove duplicates
temp_director_df.drop_duplicates(["title", "director", 'last name'], keep='last', inplace=True)
temp_director_df.columns = ['title', 'first_name', 'last_name']

#add index
temp_director_df['director_id'] = [i for i in range(1, len(temp_director_df) + 1)]

# drop columns
new_director_df = temp_director_df.loc[:, ['first_name', 'last_name']]
new_director_df.columns = ['first_name', 'last_name']


#remove duplicates
new_director_df.drop_duplicates(['first_name', 'last_name'], keep='last', inplace=True)

#add index
new_director_df['director_id'] = [i for i in range(1, len(new_director_df) + 1)]
new_director_df = new_director_df[['director_id', 'first_name', 'last_name']]


new_director_df
```

### Table five: Movie Directors table

```python
#merge tables
director_df_join = pd.merge(temp_director_df, 
                            new_director_df,
                            on = ['first_name', 'last_name'])

disney_director_df = pd.merge(new_movie_df, 
                              director_df_join,
                              left_on = "movie_title",
                              right_on = 'title')
```

```python
# build movie director table
movie_director_df = disney_director_df.copy()
movie_director_df = movie_director_df[['movie_id', 'director_id_y']]
movie_director_df.columns = ['movie_id', 'director_id']
movie_director_df
```

```python
#top 10 movies with most directors
director_grouped = disney_director_df.groupby(['title']).count()[['first_name']].sort_values(by=['first_name'], ascending=False)
director_grouped.columns = ['Number of Directors']
director_grouped.head(10)
```

```python
# top 10 directors that direct most disney movies
director_count = disney_director_df.groupby(['first_name', 'last_name']).count()[['title']].sort_values(by="title", ascending=False)
director_count.columns = ['Number of Disney Movies']
director_count.head(10)
```

### Table six: Actor table

```python
temp_actor_df = actor_df.copy()

# split actor's name
temp_actor_df.loc[temp_actor_df['cast'].str.split().str.len() == 2, 'last name'] = temp_actor_df['cast'].str.split().str[-1]

# dropna
temp_actor_df = temp_actor_df[temp_actor_df['last name'].notna()]

# remove duplicates
temp_actor_df.drop_duplicates(["title", "cast", "last name"], keep='last', inplace=True)
# rename columns
temp_actor_df.columns = ['title', 'first_name', 'last_name']

# add index
temp_actor_df['actor_id'] = [i for i in range(1, len(temp_actor_df) + 1)]


# drop columns
new_actor_df = temp_actor_df.loc[:, ['first_name', 'last_name']]
new_actor_df.columns = ['first_name', 'last_name']

# remove duplicates
new_actor_df.drop_duplicates(['first_name', 'last_name'], keep='last', inplace=True)

# add index
new_actor_df['actor_id'] = [i for i in range(1, len(new_actor_df) + 1)]
new_actor_df = new_actor_df[['actor_id', 'first_name', 'last_name']]

new_actor_df
```

### Table seven: Movie Actor table


```python
#merge tables
actor_df_join = pd.merge(temp_actor_df, 
                        new_actor_df,
                        on=['first_name', 'last_name'])

disney_actor_df = pd.merge(new_movie_df, 
                           actor_df_join,
                           left_on = "movie_title",
                           right_on= 'title')

disney_actor_df.head()
```

```python
# build Movie Actor table
movie_actor_df = disney_actor_df.copy()
# select columns
movie_actor_df = movie_actor_df[['movie_id','actor_id_y']]
# rename columns
movie_actor_df.columns = ['movie_id', 'actor_id']
movie_actor_df
```

```python
#top 10 movies with most actors
movie_actors = disney_actor_df.groupby(['title']).count()[['first_name']].sort_values(by=['first_name'], ascending=False)
movie_actors.columns = ['Number of Actors on the list']
movie_actors.head(10)
```

```python
# top 10 directors that direct most disney movies
actors_count = disney_actor_df.groupby(['first_name', 'last_name']).count()[['title']].sort_values(by="title", ascending=False)
actors_count.columns = ['Number of Disney Movies']
actors_count.head(10)
```

# Step 3: Load
* Note: Once the database is ready, the ETL process can run. 
* SQL schema: https://github.com/wiseqx/ETL_Project/blob/master/sql/SQL_Schema.sql


```python
# connect to database
# note: put your local path here
engine = create_engine('postgresql://postgres/disney_db')
connection = engine.connect()
```

```python
# load all of data to PostgresSQL database
box_df.to_sql(name='boxoffice', con=engine, if_exists='append', index=False)
new_movie_df.to_sql(name='movies', con=engine, if_exists='append', index=False)
new_rating_df.to_sql(name='reviews', con=engine, if_exists='append', index=False)
new_director_df.to_sql(name='directors', con=engine, if_exists='append', index=False)
movie_director_df.to_sql(name='movie_directors', con=engine, if_exists='append', index=False)
new_actor_df.to_sql(name='voice_actors', con=engine, if_exists='append', index=False)
movie_actor_df.to_sql(name='movie_actors', con=engine, if_exists='append', index=False)
```
